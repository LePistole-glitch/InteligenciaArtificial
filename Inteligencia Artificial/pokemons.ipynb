```python
df = pd.read_csv(r"C:\Users\chave\OneDrive\Escritorio\9no Semestre\Inteligencia Artificial\archive\pokemons.csv")
df.head(5)
```

```python
df['type1'].value_counts().plot(kind='bar')
plt.xlabel('Tipo de Pokémon')
plt.ylabel('Número de Pokémon')
plt.title('Distribución de Tipo1 de Pokémon')
plt.show()
```

```python
df['type2'].value_counts().plot(kind='bar')
plt.xlabel('Tipo de Pokémon')
plt.ylabel('Número de Pokémon')
plt.title('Distribución de Tipo2 de Pokémon')
plt.show()
```

```python
pokemon_stats = df.iloc[0]
stats = ['hp', 'atk', 'def', 'spatk', 'spdef', 'speed']
stats_values = pokemon_stats[stats]
angles = [n / float(len(stats)) * 2 * 3.14159 for n in range(len(stats))]
angles += angles[:1]
stats_values = stats_values.tolist()
stats_values += stats_values[:1]
plt.polar(angles, stats_values)
plt.fill(angles, stats_values, 'b', alpha=0.1)
plt.xticks(angles[:-1], stats)
plt.title('Estadísticas de un Pokémon en Forma de Radar')
plt.show()
```

```python
gen_counts = df['generation'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(gen_counts, labels=gen_counts.index)
plt.title('Distribución de Pokémon por Generación!!')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['atk'], df['def'])
plt.xlabel('Ataque')
plt.ylabel('Defensa')
plt.title('Relación entre Ataque y Defensa de Pokémon')
plt.grid(True)
plt.show()
```

```python
type1_counts = df['type1'].value_counts()
type1_counts.plot(kind='pie')
plt.axis('equal')
plt.title('Distribución de Tipos Primarios')
plt.show()
```

```python
top_10_atk = df.nlargest(10, 'atk')

plt.figure(figsize=(10, 6))
plt.barh(top_10_atk['name'], top_10_atk['atk'])
plt.xlabel('Ataque')
plt.ylabel('Nombre del Pokémon')
plt.title('Top 10 Pokémon con las Estadísticas de Ataque más Altas')
plt.gca().invert_yaxis()
plt.show()
```

```python
top_10_def = df.nlargest(10, 'def')
plt.figure(figsize=(10, 6))
plt.barh(top_10_def['name'], top_10_def['def'])
plt.xlabel('Defensa')
plt.ylabel('Nombre del Pokémon')
plt.title('Top 10 Pokémon con las Estadísticas de Defensa más Altas')
plt.gca().invert_yaxis()
plt.show()
```

```python
pokemon_names = ['pikachu', 'bulbasaur', 'charmander', 'squirtle']

selected_pokemon = df[df['name'].isin(pokemon_names)]

selected_pokemon = selected_pokemon.set_index('name')
selected_pokemon = selected_pokemon[['hp', 'atk', 'def', 'spatk', 'spdef', 'speed']].T

selected_pokemon.plot(kind='line', marker='o', figsize=(10, 6))
plt.xlabel('Estadísticas')
plt.ylabel('Valor')
plt.title('Estadísticas de Varios Pokémon Iniciales')
plt.legend(title='Pokémon', loc='upper left')
plt.show()
```

```python
gen_counts = df['rank'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(gen_counts, labels=gen_counts.index)
plt.title('Distribución de Pokémon por rango')
plt.show()
```
